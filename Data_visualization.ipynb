```python
# Import libraries
import matplotlib.pyplot as plt
from numpy import random
import seaborn as sns


years = range(2012,2022)
apples = random.rand(10)
oranges = random.rand(10)
```

```python
# use seaborn to change plot whitegrid/darkgrid
sns.set_style("whitegrid")
```

```python
#Line Plot - Matplotlib
plt.plot (years,apples)
plt.plot (years,oranges,marker ='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons/ hectare)')
plt.title('Crop yields in Australia')
plt.legend(['Apples', 'Oranges'])
```

```python
#Bar Plot - Matplotlib
plt.bar (years,apples)
#plt.plot (years,oranges,marker ='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons/ hectare)')
plt.title('Crop yields in Australia')
#plt.legend(['Apples', 'Oranges'])
```

```python
#Histogram Plot - Matplotlib
plt.hist (apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons/ hectare)')
plt.title('Crop yields in Australia')

```

```python
sns.scatterplot(x=years,y=apples)
```

```python
sns.heatmap(apples[:, np.newaxis])
```
